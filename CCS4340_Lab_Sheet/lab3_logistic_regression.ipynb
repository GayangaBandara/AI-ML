Import Required Libraries

```python
import numpy as np
import pandas as pd 

#to visualizations 
import matplotlib.pyplot as plt

#to divide data set into training and test set
from sklearn.model_selection import train_test_split

#to build the logistic regression model
from sklearn.linear_model import LogisticRegression
```

Load the Dataset

```python
data = pd.read_csv('..\\Dataset\\Social_Network_Ads.csv')
```

Explore the Dataset

```python
#see first 5 rows of the dataset
data.head()
```

Visualize the Data

```python
plt.scatter(data.Age, data.Purchased)
```

```python
plt.scatter(data.EstimatedSalary,data.Purchased)
```

Prepare Features and Labels

```python
#see a column from the dataset
x = data[["Age","EstimatedSalary"]]
x
```

```python
y = data['Purchased']
y
```

```python
#divide dataset
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_test
```

Train the Logistic Regression Model

```python
#build model
model = LogisticRegression()
```

```python
#train model
model.fit(x_train,y_train)
```

Make Predictions

```python
#get predictions form model
model.predict(x_test)
```

```python
#accuracy of the model
model.score(x_test,y_test)
```

```python
new_values = np.array([[20,10000],[25,15000],[40,50000]])
feature_names = ["Age","EstimatedSalary"]
new_data = pd.DataFrame(new_values,columns=feature_names)
new_data
```
